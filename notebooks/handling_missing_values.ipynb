## Handling missing values

### Price and Price_R

We are just keeping price in €

```python
df = df.drop(columns='price_R')
```

```python
df[df['price'].isna()]
```

As price is the target variable, we will handle missing values by imputing the

```python
df['price'] = df.groupby(['neighbourhood_cleansed', 'room_type'])['price'].transform(lambda x: x.fillna(x.mean()))
```

```python
sum(df['price'].isna())
```

```python
df['price'] = df['price'].fillna(df['price'].median())
```

```python
sum(df['price'].isna())
```

### Last review

```python
df[df['last_review'].isna()]
```

```python
df = df.dropna(subset=['last_review'])
```

### Host response time and rate

```python
df[df['host_response_time'].isna()]
```

```python
df['host_response_rate'].mean()
```

```python
df['host_response_time'].mode()
```

```python
df['host_response_rate'] = df['host_response_rate'].fillna(df['host_response_rate'].mean())
df['host_response_time'] = df['host_response_time'].fillna(df['host_response_time'].mode()[0])
```

### Host is superhost

```python
df[df['host_is_superhost'].isna()]
```

```python
df['host_is_superhost'].value_counts()
```

```python
df['host_is_superhost'] = df['host_is_superhost'].fillna(df['host_is_superhost'].mode()[0])
```

```python
df['host_is_superhost'].isna().sum()
```

### Review scores

```python
df[df['review_scores_value'].isna()]
```

```python
df = df.dropna(subset=['review_scores_value'])
```

```python
df = df.dropna(subset=['review_scores_checkin'])
```

```python
df = df.dropna(subset=['review_scores_cleanliness'])
```

### Host acceptance rate

```python
df[df['host_acceptance_rate'].isna()]
```

```python
df['host_acceptance_rate'].mean()
```

```python
df['host_acceptance_rate'] = df.groupby('host_response_rate')['host_acceptance_rate'].transform(lambda x: x.fillna(x.median()))
```
